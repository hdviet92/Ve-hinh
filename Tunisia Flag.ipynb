```python
# Bước 1: Import
import turtle

# Bước 2: Tọa đối tượng turtle
tl= turtle.Turtle()
tl.shape('turtle')
tl.hideturtle()
tl.speed(5)

# Bước 3: Control tl
tl.penup()
tl.goto(120,80)
tl.pendown()
tl.pencolor('red')
tl.fillcolor('red')
tl.begin_fill()
tl.right(90)
tl.forward(160)
tl.right(90)
tl.forward(240)
tl.right(90)
tl.forward(160)
tl.right(90)
tl.forward(240)
tl.end_fill()

tl.penup()
tl.goto(0,-40)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
tl.circle(40)
tl.end_fill()

tl.penup()
tl.goto(0,-30)
tl.pendown()
tl.fillcolor('red')
tl.begin_fill()
tl.circle(30)
tl.end_fill()

tl.penup()
tl.goto(8,-24)
tl.pendown()
tl.pencolor('white')
tl.fillcolor('white')
tl.begin_fill()
tl.circle(24)
tl.end_fill()

tl.penup()
tl.goto(-10,0)
tl.pendown()
tl.pencolor('red')
tl.fillcolor('red')
tl.begin_fill()
tl.left(18)
for x in range(5):
    tl.forward(36)
    tl.right(144)
tl.end_fill()

# Bước 4: End
turtle.done()

```
