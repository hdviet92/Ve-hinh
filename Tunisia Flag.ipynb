```python
import turtle
# KHAI BÁO BIẾN
wi = 60*10
he = 40*10

# Tọa đối tượng turtle
tl = turtle.Turtle()
tl.shape('turtle')
tl.hideturtle()
tl.speed(5)

# Vẽ khung ngoài
tl.penup()
tl.goto(wi/2,he/2)
tl.pendown()
tl.pencolor('red')
tl.fillcolor('red')
tl.begin_fill()
tl.right(90)
tl.forward(he)
tl.right(90)
tl.forward(wi)
tl.right(90)
tl.forward(he)
tl.right(90)
tl.forward(wi)
tl.end_fill()

# Vẽ hình tròn trắng
tl.penup()
tl.goto(0,-he/4)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
tl.circle(he/4)
tl.end_fill()

# vẽ hình tRANG KHUYỂT
tl.penup()
tl.goto(0,-wi/8)
tl.pendown()
tl.fillcolor('red')
tl.begin_fill()
tl.circle(wi/8)
tl.end_fill()

tl.penup()
tl.goto(wi/30,-wi/10)
tl.pendown()
tl.pencolor('white')
tl.fillcolor('white')
tl.begin_fill()
tl.circle(wi/10)
tl.end_fill()

# VẼ NGÔI SAO
tl.penup()
tl.goto(-he/16,0)
tl.pendown()
tl.pencolor('red')
tl.fillcolor('red')
tl.begin_fill()
tl.left(18)
for x in range(5):
    tl.forward(9*he/40)
    tl.right(144)
tl.end_fill()

# Bước 4: End
turtle.done()

```
