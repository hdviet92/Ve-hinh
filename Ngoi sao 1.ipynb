```python
# Bước 1: Import module turtle
import turtle

# Bước 2: Tạo đối tượng turtle
tl = turtle.Turtle()
tl.shape("turtle")
tl.pensize(3)
tl.pencolor('yellow')
tl.speed(1)
#tl.fillcolor('orange')

#  Bước 3: Điều khiển đối tượng tl
tl.right('60')
tl.forward('200')
tl.right('120')
tl.forward('200')
tl.right('120')
tl.forward('200')
tl.right('120')
tl.forward('200')
tl.right('120')
tl.forward('200')

# Bước 4: Hoàn thành
turtle.done()
```
