```python

import turtle
x = 100

tl = turtle.Turtle()
tl.hideturtle()
tl.speed(9)

# than
tl.circle(-x)
tl.circle(x/1.5)

# khan
tl.penup()
tl.goto(x/2,x/10)
tl.pendown()
tl.pencolor('orange')
tl.fillcolor('orange')
tl.begin_fill()
tl.backward(x)
tl.right(90)
tl.forward(x/4)
tl.left(90)
tl.forward(x)
tl.left(90)
tl.forward(x/4)
tl.end_fill()

# mui
tl.penup()
tl.goto(x/8,x/1.7)
tl.pendown()
tl.pencolor('red')
tl.fillcolor('red')
tl.begin_fill()
tl.circle(x/8,360,6)
tl.end_fill()

# mat
tl.penup()
tl.goto(x/1.9,x/1.2)
tl.pendown()
tl.pencolor('blue')
tl.fillcolor('blue')
tl.begin_fill()
tl.circle(x/12)
tl.penup()
tl.goto(-x/1.9,x/1.2)
tl.pendown()
tl.circle(-x/12)
tl.end_fill()

# tay
tl.penup()
tl.goto(-x/1.2,-x/2)
tl.pendown()
tl.pensize(15)
tl.pencolor('brown')
tl.fillcolor('brown')
tl.begin_fill()
tl.left(30)
tl.forward(1.5*x)
tl.end_fill()
tl.penup()
tl.goto(x/1.2,-x/2)
tl.pendown()
tl.pencolor('brown')
tl.fillcolor('brown')
tl.right(60)
tl.forward(1.5*x)
tl.end_fill()



turtle.done()
```
