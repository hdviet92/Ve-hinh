```python

import turtle
x = 100

tl = turtle.Turtle()
#tl.hideturtle()
tl.speed(20)

# tay phai
tl.pencolor('brown')
tl.pensize(x/16)
tl.penup()
tl.goto(-x/1.5,-x/2)
tl.pendown()
tl.left(150)
tl.forward(1.5*x)

# than
tl.penup()
tl.goto(0,0)
tl.pendown()
tl.pensize(1)
tl.right(150)
tl.fillcolor('white')
tl.begin_fill()
tl.circle(-x)
tl.circle(x/1.5)
tl.end_fill()

# mui
tl.penup()
tl.goto(0,x/1.23)
tl.pendown()
tl.pencolor('orange')
tl.fillcolor('orange')
tl.begin_fill()
tl.circle(-x/12.4)
tl.goto(0,x/1.3)
tl.right(90)
tl.forward(x/8)
tl.right(90)
tl.forward(x/1.6)
tl.right(165)
tl.forward(x/1.5)
tl.end_fill()

# mat
tl.right(15)
tl.penup()
tl.goto(x/4,x/1.3)
tl.pendown()
tl.pencolor('black')
tl.fillcolor('black')
tl.begin_fill()
tl.circle(x/16)
tl.end_fill()
tl.left(90)
tl.forward(x/16)
tl.left(90)
tl.forward(x/16)
tl.left(90)
tl.fillcolor('white')
tl.begin_fill()
tl.circle(x/22)
tl.end_fill()
tl.left(90)

tl.penup()
tl.goto(-x/3,x/1.3)
tl.pendown()
tl.pencolor('black')
tl.fillcolor('black')
tl.begin_fill()
tl.circle(x/16)
tl.end_fill()
tl.left(90)
tl.forward(x/16)
tl.left(90)
tl.forward(x/16)
tl.left(90)
tl.fillcolor('white')
tl.begin_fill()
tl.circle(x/22)
tl.end_fill()
tl.left(90)

# mieng
tl.penup()
tl.goto(0,x/3)
tl.pendown()
tl.pencolor('black')
tl.pensize(x/50)
tl.circle(x/3,50)
tl.left(100)
tl.forward(x/16)
tl.backward(x/8)

# tay phai
tl.pencolor('brown')
tl.pensize(x/16)
tl.penup()
tl.goto(x/1.5,-x/2)
tl.pendown()
tl.right(120)
tl.forward(1.5*x)

# khan
tl.penup()
tl.goto(x/2.2,0)
tl.pendown()
tl.right(30)
tl.pencolor('red')
tl.pensize(x/3)
tl.backward(x/1.1)
tl.forward(x/1.4)
tl.pensize(x/5)
tl.right(80)
tl.forward(x/1.5)
tl.backward(x/1.5)
tl.right(30)
tl.forward(x/2)

# nut
tl.penup()
tl.goto(-x/4,-x/2)
tl.pendown()
tl.pensize(1)
tl.pencolor("red")
tl.fillcolor('red')
tl.begin_fill()
tl.circle(x/11)
tl.end_fill()
tl.goto(-10*x/44,-x/1.95)
tl.fillcolor('yellow')
tl.begin_fill()
tl.circle(x/22)
tl.end_fill()

tl.penup()
tl.goto(-x/4,-x)
tl.pendown()
tl.pencolor("red")
tl.fillcolor('red')
tl.begin_fill()
tl.circle(x/11)
tl.end_fill()
tl.goto(-10*x/44,-x/0.99)
tl.fillcolor('yellow')
tl.begin_fill()
tl.circle(x/22)
tl.end_fill()

# mu
tl.left(110)
tl.penup()
tl.goto(0,2*x/1.7)
tl.left(180)
tl.pendown()
tl.pencolor("green")
tl.circle(x,30)
tl.right(130)
tl.forward(x/4)
tl.right(50)
tl.circle(-x,30)

tl.penup()
tl.goto(0,2*x/1.7)
tl.pendown()
tl.pencolor("green")
tl.circle(-x,30)
tl.left(130)
tl.forward(x/4)
tl.left(50)
tl.circle(x,30)



# end
turtle.done()

```
